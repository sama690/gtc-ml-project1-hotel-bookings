```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(x=df["adr"], ax=axes[0])
axes[0].set_title("ADR")
sns.boxplot(x=df["lead_time"], ax=axes[1])
axes[1].set_title("Lead Time")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))
#adr
Q1 = df["adr"].quantile(0.25)
Q3 = df["adr"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
outliers_adr = df[(df["adr"] < lower) | (df["adr"] > upper)]
sns.boxplot(x=df["adr"], showfliers=False, color="skyblue", ax=axes[0])
axes[0].scatter(outliers_adr["adr"], [0]*len(outliers_adr), color="red", s=20, label="Outliers")
axes[0].legend()
axes[0].set_title("ADR Outliers")
axes[0].text(0.5, -0.25,f"Q1={Q1:.2f}, Q3={Q3:.2f}\nLower={lower:.2f}, Upper={upper:.2f}\nOutliers={outliers_adr.shape[0]}",ha="center", va="top", transform=axes[0].transAxes)
#lead_time
Q1 = df["lead_time"].quantile(0.25)
Q3 = df["lead_time"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
outliers_lead = df[(df["lead_time"] < lower) | (df["lead_time"] > upper)]
sns.boxplot(x=df["lead_time"], showfliers=False, color="skyblue", ax=axes[1])
axes[1].scatter(outliers_lead["lead_time"], [0]*len(outliers_lead), color="red", s=20, label="Outliers")
axes[1].legend()
axes[1].set_title("Lead Time Outliers")
axes[1].text(0.5, -0.25,f"Q1={Q1:.2f}, Q3={Q3:.2f}\nLower={lower:.2f}, Upper={upper:.2f}\nOutliers={outliers_lead.shape[0]}",ha="center", va="top", transform=axes[1].transAxes)
plt.tight_layout()
plt.show()
```

```python
# Data Quality Issues

# Several columns contain missing values (agent, company,country).
# Some columns have incorrect data types (dates stored as strings).
# Detected outliers in `adr` and `lead_time` using IQR and boxplots.
# Found one negative value in `adr`, which is not logically valid.
# Potential data entry inconsistencies to be checked further.
```

```python
#Phase2
missing_values_count=df.isnull().sum()
total_cells = np.prod(df.shape)
total_missing =missing_values_count.sum()
percent_missing = (total_missing/total_cells) * 100
print(percent_missing,"%")
```

```python
#Handle Missing Values
df_copy = df.copy()
df_copy[["company", "agent"]] = df_copy[["company", "agent"]].fillna(0)

df_copy["country"] = df_copy["country"].fillna("Unknown")

median_children = df_copy["children"].median()
df_copy["children"] = df_copy["children"].fillna(median_children)
```

```python
print("Duplicated rows:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
#Handle Outliers
df_copy['adr'] = df_copy['adr'].apply(lambda x: 1000 if x > 1000 else x)
plt.figure(figsize=(10,4))
sns.boxplot(x=df_copy['adr'], color='skyblue')
plt.title("ADR Boxplot After Capping at 1000")
plt.xlabel("ADR")

plt.show()
# Capping extreme values in 'adr' to handle outliers
# Justification:
# ADR has some extremely high values (e.g., 5000+) which can skew the analysis and models.
# By capping values at 1000, we preserve most of the data while limiting the influence of outliers.
```

```python
# create a new column, date_parsed, with the parsed dates
df['date_parsed'] = pd.to_datetime(df['reservation_status_date'], format="%Y-%m-%d")
```

```python
#Feature Engineering & Preprocessing
# Total guests
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Total nights
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# Is family
df['is_family'] = df.apply(
    lambda row: "Yes" if (row['children'] + row['babies']) > 0 else "No", axis=1
)
```

```python
df.drop('reservation_status_date', axis=1, inplace=True)
```

```python
df.drop('reservation_status', axis=1, inplace=True)
```

```python
low_cardinality = ['meal','market_segment', 'distribution_channel']
df= pd.get_dummies(df , columns= low_cardinality, drop_first= True)
```

```python
threshold = 100
country_counts = df_copy['country'].value_counts()
countries_to_keep = country_counts[country_counts >= threshold].index

df_copy['country_grouped'] = df_copy['country'].apply(
    lambda x: x if x in countries_to_keep else "Other"
)
# Frequency encoding
freq_encoding = df_copy['country_grouped'].value_counts(normalize=True)
df_copy['country_encoded'] = df_copy['country_grouped'].map(freq_encoding)
df_copy.drop('country', axis=1, inplace=True)
```
